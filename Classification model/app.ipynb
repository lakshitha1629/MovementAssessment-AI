# Model Training

```python
from sklearn.model_selection import train_test_split
# train - 70
# test - 30
X = df.drop(columns=['Output'])
Y = df['Output']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.30)
```

```python
from sklearn.tree import DecisionTreeClassifier
# model = LogisticRegression()
# model = LinearDiscriminantAnalysis()
model = KNeighborsClassifier()
# model = DecisionTreeClassifier()
# model = GaussianNB()
# model = SVC()
# model = GradientBoostingClassifier(n_estimators=20, learning_rate=0.5, max_features=2, max_depth=2, random_state=0)

# model training
model.fit(x_train, y_train)
```

```python
print("Accuracy: ",model.score(x_test, y_test) * 100)
```

```python
def predict_mpg(config, model):
    
    if type(config) == dict:
        df = pd.DataFrame(config)
    else:
        df = config
    
#     print(prepared_df)
    y_pred = model.predict(df)
    return y_pred
    
```

```python
config = {
    'Collar_bone_x': [-0.1559136],
    'Collar_bone_y': [0.2286534],
    'Collar_bone_z': [2.120145],
    'Fore_arm_x': [-0.5302216],
    'Fore_arm_y': [0.2827062],
    'Fore_arm_z': [2.109175],
    'Hand_x_test': [-0.5868202],
    'Hand_y_test': [0.2925405],
    'Hand_z_test': [2.103178],
    'Upper_arm_x': [-0.3435675],
    'Upper_arm_y': [0.2677751],
    'Upper_arm_z': [2.130795]
    
}

predict_mpg(config, model)
```
